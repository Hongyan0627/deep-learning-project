```python

# VGG19
import pandas as pd
import numpy as np 
import pylab as pl 
from sklearn.metrics import roc_curve, auc
print 'come into approach 1'
df_1 = pd.read_csv('stage1_solution.csv')
# df_2 approach 1
df_2 = pd.read_csv('submitvgg19.csv')
# merge_1 approach 1
merge_1 = df_1.merge(df_2, on = 'id')
# print 'merge csv', merge
keep_cols = ['cancer_x', 'cancer_y']
# df_a1 approach 1
df_a1 = merge_1[keep_cols]
# print 'finish reading csv file'
y_test_a1 = np.array(df_a1)[:,0]
probas_a1 = np.array(df_a1)[:,1]
# print y_test
# print 'probas', probas
# Compute ROC curve and area the curve
fpr_1, tpr_1, thresholds_1 = roc_curve(y_test_a1, probas_a1)
# print 'thresholds', thresholds
roc_auc_1 = auc(fpr_1, tpr_1)
print("Area under the ROC curve : %f" % roc_auc_1)
print 'finish approach 1'

print 'come into approach 2 avg'
df_3 = pd.read_csv('submitvgg19_label_image_Avg.csv')
# merge_1 approach 1
merge_avg = df_1.merge(df_3, on = 'id')
# print 'merge csv', merge
keep_cols = ['cancer_x', 'cancer_y']
# df_a1 approach 1
df_a2_avg = merge_avg[keep_cols]
# print 'finish reading csv file'
y_test_a2_avg = np.array(df_a2_avg)[:,0]
probas_a2_avg = np.array(df_a2_avg)[:,1]
# print y_test
# print 'probas', probas
# Compute ROC curve and area the curve
fpr_a2_avg, tpr_a2_avg, thresholds_a2_avg = roc_curve(y_test_a2_avg, probas_a2_avg)
# print 'thresholds', thresholds
roc_auc_a2_avg = auc(fpr_a2_avg, tpr_a2_avg)
print("Area under the ROC curve : %f" % roc_auc_a2_avg)
print 'finish approach 2 avg'

print 'come into approach 2 max'
df_4 = pd.read_csv('submitvgg19_label_image_Max.csv')
# merge_1 approach 1
merge_max = df_1.merge(df_4, on = 'id')
# print 'merge csv', merge
keep_cols = ['cancer_x', 'cancer_y']
# df_a1 approach 1
df_a2_max = merge_max[keep_cols]
# print 'finish reading csv file'
y_test_a2_max = np.array(df_a2_max)[:,0]
probas_a2_max = np.array(df_a2_max)[:,1]
# print y_test
# print 'probas', probas
# Compute ROC curve and area the curve
fpr_a2_max, tpr_a2_max, thresholds_a2_max = roc_curve(y_test_a2_max, probas_a2_max)
# print 'thresholds', thresholds
roc_auc_a2_max = auc(fpr_a2_max, tpr_a2_max)
print("Area under the ROC curve : %f" % roc_auc_a2_max)
print 'finish approach 2 max'

print 'come into approach 2 min'
df_5 = pd.read_csv('submitvgg19_label_image_Min.csv')
# merge_1 approach 1
merge_min = df_1.merge(df_5, on = 'id')
# print 'merge csv', merge
keep_cols = ['cancer_x', 'cancer_y']
# df_a1 approach 1
df_a2_min = merge_min[keep_cols]
# print 'finish reading csv file'
y_test_a2_min = np.array(df_a2_min)[:,0]
probas_a2_min = np.array(df_a2_min)[:,1]
# print y_test
# print 'probas', probas
# Compute ROC curve and area the curve
fpr_a2_min, tpr_a2_min, thresholds_a2_min = roc_curve(y_test_a2_min, probas_a2_min)
# print 'thresholds', thresholds
roc_auc_a2_min = auc(fpr_a2_min, tpr_a2_min)
print("Area under the ROC curve : %f" % roc_auc_a2_min)
print 'finish approach 2 min'

pl.clf()
pl.plot(fpr_1, tpr_1, label='vgg19 ROC curve (area = %0.2f)' % roc_auc_1)
pl.plot(fpr_a2_avg, tpr_a2_avg, label = 'vgg19 a2 avg roc curve (area = %0.2f)' % roc_auc_a2_avg)
pl.plot(fpr_a2_max, tpr_a2_max, label = 'vgg19 a2 max roc curve (area = %0.2f)' % roc_auc_a2_max)
pl.plot(fpr_a2_min, tpr_a2_min, label = 'vgg19 a2 min roc curve (area = %0.2f)' % roc_auc_a2_min)
pl.plot([0, 1], [0, 1], 'k--')
pl.xlim([0.0, 1.0])
pl.ylim([0.0, 1.0])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('VGG19 ROC comparison')
pl.legend(loc="lower right")
pl.show()

from sklearn import metrics
print 'vgg19 a1 loss', (metrics.log_loss(y_test_a1, probas_a1))
print 'vgg19 a2 avg loss',(metrics.log_loss(y_test_a2_avg, probas_a2_avg))
print 'vgg19 a2 max loss',(metrics.log_loss(y_test_a2_max, probas_a2_max))
print 'vgg19 a2 min loss',(metrics.log_loss(y_test_a2_min, probas_a2_min))
```

```python
# Compare between VGG19 and ResNet50 within their best approach
import pandas as pd
import numpy as np 
import pylab as pl 
from sklearn.metrics import roc_curve, auc
print 'come into approach resNet'
df_1 = pd.read_csv('stage1_solution.csv')
# df_2 approach 1
df_resNet = pd.read_csv('submitresNet.csv')
# merge_1 approach 1
merge_resNet = df_1.merge(df_resNet, on = 'id')
# print 'merge csv', merge
keep_cols = ['cancer_x', 'cancer_y']
# df_a1 approach 1
df_resNet = merge_resNet[keep_cols]
# print 'finish reading csv file'
y_test_resNet = np.array(df_resNet)[:,0]
probas_resNet = np.array(df_resNet)[:,1]
# print y_test
# print 'probas', probas
# Compute ROC curve and area the curve
fpr_resNet, tpr_resNet, thresholds_resNet = roc_curve(y_test_resNet, probas_resNet)
# print 'thresholds', thresholds
roc_auc_resNet = auc(fpr_resNet, tpr_resNet)
print("Area under the ROC curve : %f" % roc_auc_resNet)
print 'finish approach resNet'

print 'come into approach vgg19'
df_vgg19 = pd.read_csv('submitvgg19.csv')

merge_vgg19 = df_1.merge(df_vgg19, on = 'id')

keep_cols = ['cancer_x', 'cancer_y']

df_vgg19 = merge_vgg19[keep_cols]

y_test_vgg19 = np.array(df_vgg19)[:,0]
probas_vgg19 = np.array(df_vgg19)[:,1]
fpr_vgg19, tpr_vgg19, thresholds_vgg19 = roc_curve(y_test_vgg19, probas_vgg19)
# print 'thresholds', thresholds
roc_auc_vgg19 = auc(fpr_vgg19, tpr_vgg19)
print("Area under the ROC curve : %f" % roc_auc_vgg19)
print 'finish approach vgg19'

print 'comes into combine two models'
df_combine = pd.read_csv('submitresNet_vgg.csv')
merge_combine = df_1.merge(df_combine, on = 'id')
print 'merge csv', merge_combine
keep_cols = ['cancer_x', 'cancer_y']
df = merge_combine[keep_cols]
print 'finish reading csv file'
y_test_combine = np.array(df)[:,0]
probas_combine = np.array(df)[:,1]

fpr_combine, tpr_combine, thresholds_combine = roc_curve(y_test_combine, probas_combine)
# print 'thresholds', thresholds_combine
roc_auc_combine = auc(fpr_combine, tpr_combine)
print("Area under the ROC curve : %f" % roc_auc_combine)

pl.clf()
pl.plot(fpr_resNet, tpr_resNet, label='resNet50 ROC curve (area = %0.2f)' % roc_auc_resNet)
pl.plot(fpr_vgg19, tpr_vgg19, label='vgg19 ROC curve (area = %0.2f)' % roc_auc_vgg19)
pl.plot(fpr_combine, tpr_combine, label='two model combine ROC curve (area = %0.2f)' % roc_auc_combine)
pl.plot([0, 1], [0, 1], 'k--')
pl.xlim([0.0, 1.0])
pl.ylim([0.0, 1.0])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('ResNet50 and VGG19 ROC curve comparison')
pl.legend(loc="lower right")
pl.show()

from sklearn import metrics
print 'resNet loss', (metrics.log_loss(y_test_resNet, probas_resNet))
print 'vgg19 loss', (metrics.log_loss(y_test_vgg19, probas_vgg19))
print 'combine loss', (metrics.log_loss(y_test_combine, probas_combine))
```
